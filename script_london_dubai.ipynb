## Outliers

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sélection des colonnes numériques
numeric_cols = df.select_dtypes(include=['number']).columns

# Définir la taille de la figure pour les boxplots
plt.figure(figsize=(15, len(numeric_cols) * 2.5))

# Générer les boxplots
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(len(numeric_cols), 1, i)
    sns.boxplot(x=df[col], orient='h')
    plt.title(f'Boxplot de {col}')
    plt.tight_layout()

plt.show()
```

## Distribution des variables

```python
numeric_columns = df.select_dtypes(include=np.number).columns

# Définir la taille de la figure
n_cols = 3
n_rows = int(np.ceil(len(numeric_columns) / n_cols))
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4))
axes = axes.flatten()

# Créer un histogramme avec courbe de densité pour chaque variable
for i, col in enumerate(numeric_columns):
    sns.histplot(df[col], kde=True, ax=axes[i], bins=30)
    axes[i].set_title(f'Distribution de {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Fréquence')

# Supprimer les axes vides si le nombre de variables < nombre de sous-graphiques
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

# Tests statistiques

```python
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# ANOVA : comparaison des moyennes d'AQI selon les villes
model = ols('AQI ~ C(City)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

p-value < 0.05 → On rejette l’hypothèse nulle, donc au moins une 
ville a une moyenne d’AQI différente des autres. 

p-value ≥ 0.05 → Aucune différence significative entre les 
moyennes des villes.

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Liste pour stocker les résultats
resultats_anova = []

# Sélection des colonnes de type float
colonnes_float = df.select_dtypes(include='float').columns

# Boucle sur chaque variable
for col in colonnes_float:
    try:
        model = ols(f'Q("{col}") ~ C(City)', data=df).fit()
        anova_table = sm.stats.anova_lm(model, typ=2)
        
        # Extraire la p-value pour "C(City)"
        p_value = anova_table.loc['C(City)', 'PR(>F)']
        F_stat = anova_table.loc['C(City)', 'F']
        
        # Ajouter au tableau final
        resultats_anova.append({
            'Variable': col,
            'F': round(F_stat, 2),
            'p-value': round(p_value, 5)
        })

    except Exception as e:
        print(f"Erreur avec la variable {col} : {e}")

# Convertir la liste en DataFrame
df_resultats_anova = pd.DataFrame(resultats_anova)

# Afficher le tableau complet
print(df_resultats_anova)
```
